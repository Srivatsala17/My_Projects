# Redaing the CSV files that are extracted from the Google Books API

#### We extracted two csv files. One by using API key and another one without API key. Both of them gave different number of results. So merging them and removing the duplicates. So that if there are any new records, then it can be used too.

```python
import pandas as pd
googlebooks1 = pd.read_csv('Google_Api_Dataset_01.csv')
googlebooks1.head(2)
```

```python
import pandas as pd
googlebooks2 = pd.read_csv('Google_Books_API_data.csv')
googlebooks2.head(2)
```

## Renaming the columns of both the datasets

```python


googlebooks1.rename(columns = {'volumeInfo.title' : 'title', 'volumeInfo.ISBN' : 'ISBN', 'volumeInfo.authors' : 'authors' , 'volumeInfo.publisher' : 'publisher' , 'volumeInfo.publishedDate' : 'published_date', 'volumeInfo.description' : 'description', 'volumeInfo.industryIdentifiers' : 'ISBN10/ISBN13', 'volumeInfo.pageCount':'page_count', 'volumeInfo.categories':'categories', 'volumeInfo.averageRating':'average_rating','volumeInfo.ratingsCount':'ratings_count','volumeInfo.language':'language','volumeInfo.previewLink':'preview_link','volumeInfo.imageLinks.thumbnail':'image_link','volumeInfo.maturityRating':'maturity_rating','saleInfo.isEbook':'ebook_availability', 'accessInfo.viewability':'viewability'  }, inplace = True)
```

```python


googlebooks2.rename(columns = {'volumeInfo.title' : 'title', 'volumeInfo.ISBN' : 'ISBN', 'volumeInfo.authors' : 'authors' , 'volumeInfo.publisher' : 'publisher' , 'volumeInfo.publishedDate' : 'published_date', 'volumeInfo.description' : 'description', 'volumeInfo.industryIdentifiers' : 'ISBN10/ISBN13', 'volumeInfo.pageCount':'page_count', 'volumeInfo.categories':'categories', 'volumeInfo.averageRating':'average_rating','volumeInfo.ratingsCount':'ratings_count','volumeInfo.language':'language','volumeInfo.previewLink':'preview_link','volumeInfo.imageLinks.thumbnail':'image_link','volumeInfo.maturityRating':'maturity_rating','saleInfo.isEbook':'ebook_availability', 'accessInfo.viewability':'viewability'  }, inplace = True)
```

## Changing the records of the title column to a uniform format

```python
#changing the rows under title column to a same format

googlebooks1.title = googlebooks1.title.str.title()
```

```python
#changing the rows under title column to a same format

googlebooks1.title = googlebooks1.title.str.strip()
```

```python
#Doing the same for the second dataset

googlebooks2.title = googlebooks2.title.str.title()
googlebooks2.title = googlebooks2.title.str.strip()
```

## Checking the data types of the columns before normalizing

```python
googlebooks1.info()
```

```python
googlebooks1.describe()
```

### We can see from above that the data is not normalozed

## Normalizing

```python
# normalizing teh numeric columns

col = ['average_rating', 'ratings_count']
googlebooks1[col] = (googlebooks1[col] - googlebooks1[col].min()) / (googlebooks1[col].max() - googlebooks1[col].min())
```

```python
googlebooks1.describe() 
```

```python
### We can see from above that the data is normalized now
```

```python
### Doing the same for the second dataframe
```

```python

googlebooks2.info()
```

```python
googlebooks2.describe()
```

```python
# normalizing

col = ['average_rating', 'ratings_count']
googlebooks2[col] = (googlebooks1[col] - googlebooks1[col].min()) / (googlebooks1[col].max() - googlebooks1[col].min())
```

```python
googlebooks2.describe()
```

## Merging the dataframes

```python
#now that the data is normalized merging the dataset

merged_dataset = googlebooks1.merge(googlebooks2, how='left')
```

### since both the datasets are normalized the result is also normalized

```python

merged_dataset.describe()
```

## Saving the dataframe to a file

```python

merged_dataset.to_csv('New_google_api_dataset.csv')
```
